```python
def process(image, row, col):
    collect = 0
    for i1 in range(3*row, 3*(row+1)):
        for j1 in range(3*col, 3*(col+1)):
            if image[i1,j1] == 0:
                collect = collect + 1
    if collect>=7:
        return 0
    else:
        return 255
```

```python
def image_coordinates_process(array_data):
    X_train_wfp =  np.zeros((35,35))
    X_train_1 = cv2.imread(array_data, cv2.IMREAD_GRAYSCALE)
    for i in range(0, 35):
        for j in range(0, 35):
            X_train_wfp[i,j] = process(X_train_1, i, j) 
    X_train = np.array(X_train_wfp,dtype=bool)
    X_train = np.logical_not(X_train)
    (row,col) = X_train.nonzero()
    X_train_coordinates_1 = np.array([row,col])
    X_train_coordinates_1 = np.transpose(X_train_coordinates_1)
    X_train_coordinates_1 = X_train_coordinates_1.astype(float)
    n = X_train_coordinates_1.shape[0]
    mean = np.mean(X_train_coordinates_1,axis=0)
    for i in range(n):
        X_train_coordinates_1[i,:] = X_train_coordinates_1[i,:] - mean
    if n > 120:
        n = 120
        X_train_coordinates_1 = X_train_coordinates_1[:n]
    elif n < 120:
        a = 120 - n
        n = 120
        add = np.zeros((a,2))
        for j in range(0,a):
            for k in range(0,2):
                add[j,k] = 255
        X_train_coordinates_1 = np.r_[X_train_coordinates_1, add]
    return X_train_coordinates_1
```

```python
def image_create_features(coordinate_1, coordinate2):
    features = np.zeros((120,2))
    for c_num in range(120):
        features[c_num,:] = coordinate_1[c_num,:] - coordinate2[c_num,:]
    return features
```

```python
def classification_run(folder,f_load,f_cost,ftype='cost'):
    assert ((ftype=='cost') | (ftype=='score'))
    with open(folder+'/'+fname_label) as f:
        content = f.read().splitlines()
    pairs = [line.split() for line in content]
    test_files  = [pair[0] for pair in pairs]
    train_files = [pair[1] for pair in pairs]
    answers_files = copy.copy(train_files)
    test_files.sort()
    train_files.sort()	
    ntrain = len(train_files)
    ntest = len(test_files)

    train_items = [f_load(f) for f in train_files]
    test_items  = [f_load(f) for f in test_files ]
 
    costM = np.zeros((ntest,ntrain),float)
    for i in range(ntest):
        for c in range(ntrain):
            costM[i,c] = f_cost(test_items[i],train_items[c])
    if ftype == 'cost':
        YHAT = np.argmin(costM,axis=1)
    elif ftype == 'score':
        YHAT = np.argmax(costM,axis=1)
    else:
        assert False
    correct = 0.0
    for i in range(ntest):
        if train_files[YHAT[i]] == answers_files[i]:
            correct += 1.0
    pcorrect = 100 * correct / ntest
    perror = 100 - pcorrect
    return perror
```
